```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
import sced2,scuc2,utils2,dataloader

# dataloader.load_data()
data = dataloader.load_case_data()

for day in range(1):
    for hour in range(6,12):
        for bus in data['buses']:
            bus.calculate_load(day, hour)
        for gen in data['gens']:
            gen.calculate_upper(day,hour)
        u_status = scuc2.solve_scuc(data)
        results = sced2.solve_sced(data,u_status)
        for i,gen in enumerate(data['gens']):
            gen.record_last_output(results['pg_results'][i])
```
