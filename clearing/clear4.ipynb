```python
import numpy as np
import matplotlib.pyplot as plt
import sced4,utils2,dataloader,scuc4
import scipy.io

# dataloader.load_data()
data = dataloader.load_case_data()

load_data_mat = scipy.io.loadmat('data_PROSPECT43/load_PROSPECT43.mat')
vre_data_mat = scipy.io.loadmat('data_PROSPECT43/vre_PROSPECT43.mat')

for day in range(5):

    u_status,storage_results = scuc4.solve_scuc(data,24,day)

    for hour in range(24):
        num_buses = len(data['buses'])
        node_storage_ch = [[] for _ in range(num_buses)]
        node_storage_dis = [[] for _ in range(num_buses)]
            
        for s_id, storage in enumerate(data['storages']):
            bus_idx = storage.bus_id - 1  
            node_storage_ch[bus_idx].append(storage_results['charge'][hour][s_id])
            node_storage_dis[bus_idx].append(storage_results['discharge'][hour][s_id])

        for bus in data['buses']:
            bus.calculate_load(day, hour,load_data_mat)
        for gen in data['gens']:
            gen.calculate_upper(day,hour,vre_data_mat)

        result = sced4.solve_sced(data, u_status[hour], node_storage_ch, node_storage_dis)
```
